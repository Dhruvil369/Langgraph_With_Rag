```python
from genericpath import exists
import os
import time
from typing import List, Dict
from typing_extensions import Literal
from langgraph.graph import StateGraph, START, END
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_google_genai.embeddings import GoogleGenerativeAIEmbeddings
from langchain.vectorstores import Chroma
from langchain.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.chains import RetrievalQA
from langchain.prompts import PromptTemplate
from pydantic import BaseModel
from langchain_core.messages import HumanMessage, SystemMessage
from langchain_community.tools import DuckDuckGoSearchRun

web_search = DuckDuckGoSearchRun()

# Set up API key
os.environ["GOOGLE_API_KEY"] = "AIzaSyCJaADIHnUd3TmZDfyeh2JKk_k8WO6t7JI"

# Define State
class State(Dict):
    query: str
    route: str
    response: str

# Define query step
class QuestionType(BaseModel):
    step: Literal["Banknifty", "Basics Of Stock Market", "Out of Context"]

# Initialize StateGraph
graph_builder = StateGraph(State)

# Initialize LLM
llm = ChatGoogleGenerativeAI(model="gemini-2.0-flash")
embedding_model = GoogleGenerativeAIEmbeddings(model="models/embedding-001")

#define route
router_question = llm.with_structured_output(QuestionType)


# Define prompt template
prompt_template = PromptTemplate(
    template="""
    Use the following retrieved context to answer the question:
    Context: {context}
    Question: {question}
    Answer:
    """,
    input_variables=["context", "question"]
)

# Create RetrievalQA chain
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever1,
    return_source_documents=True,
    chain_type_kwargs={"prompt": prompt_template}
)

qa_chain_bank = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever2,
    return_source_documents=True,
    chain_type_kwargs={"prompt": prompt_template}
)

def routemessage(state: State):
    start = time.time()
    response = decision = router_question.invoke([
        SystemMessage(content="""task is to determine which of the two provided PDF summaries is most relevant to a given user question.
          Output 'Banknifty' if the user question is related to following summary:
          This document focuses on various options trading strategies related to the Bank Nifty index, specifically analyzing the potential payoffs from different combinations of buying and selling put and call options.
          Key aspects covered in the document:
          Bank Nifty Price Levels: Price range from 8400 to 9800 in increments of 100.
          Options Strategies: Includes buying/selling put and call options and their respective payoffs.
          Net Payoff Calculations: The document calculates combined profits/losses of different options strategies.
          Risk Management & Decision-Making: Helps traders analyze risk-reward profiles, potential gains/losses, and breakeven points.
          Use Cases: Assists traders in choosing suitable strategies based on market conditions and risk tolerance.
          This document is relevant to questions about Bank Nifty, options trading, strategy evaluation, risk management, and financial decision-making in derivatives markets.

          Output 'Basics Of Stock Market' if the user question is related to following summary:
          This document discusses a case study involving Ajay and Venu, where Venu accepted a premium of Rs. 100,000 from Ajay. The case revolves around a speculative highway project and possible financial scenarios.
          Key aspects covered in the document:
          Ajay-Venu Transaction: The financial arrangement between the two parties.
          Speculation: The impact of news and uncertainty on financial decisions.
          Random Walk Theory: Discusses the concept and its implications.
          Options as Multi-Dimensional Instruments: Covers factors like gamma, volatility, and time decay in options trading.
          Case Studies: Used to conclude the module and provide a broader understanding of financial decision-making.
          This document is relevant to questions about case studies, speculative decision-making, financial concepts like the Random Walk Theory, and the multidimensional nature of options trading.

          Output 'Out of Context' if the user question is unrelated to the above context.

          Input Examples:
          Message: "can you explain me call options basics ?" → Output: Basics Of Stock Market
          Message: "please explain me Risk Management for Bank Nifty" → Output: Banknifty
          Message: "what is forex trading" → Output: Out of Context"""),
        HumanMessage(content=state["query"]),
    ])
    state["route"] = response.step # Treat response as a string
    end = time.time() - start
    # response_time_ms = end * 1000
    print(f"Response time for check the message is valid or not: {end} seconds")
    return {"route": state["route"]}


def retrieve_answer(state: State):
    response = qa_chain.invoke({"query": state["query"]})
    state["response"] = str(response["result"])
    print("===============================================================================")
    print(state)
    return {"response": response["result"]}

def retrieve_answer_bank(state: State):
    response = qa_chain_bank.invoke({"query": state["query"]})
    print(response)
    state["response"] = str(response["result"])
    print("===============================================================================")
    print(state)
    return {"response": response["result"]}

def retrieve_web_search(state: State):
    search_results = web_search.run(state["query"])
    state["response"] = str(search_results)
    print("===============================================================================")
    print(state)
    return {"response": search_results}

def route_decesion(state: State):
    if state["route"] == "Banknifty":
        return "Banknifty"
    elif state["route"] == "Basics Of Stock Market":
        return "Basics Of Stock Market"
    else:
        return "Out of Context"

# Add nodes
graph_builder.add_node("retrieve_answer", retrieve_answer)
graph_builder.add_node("retrieve_answer_bank", retrieve_answer_bank)
graph_builder.add_node("retrieve_web_search", retrieve_web_search)
graph_builder.add_node("routemessage", routemessage)

# Add edges
graph_builder.add_edge(START, "routemessage")
graph_builder.add_conditional_edges("routemessage", route_decesion, {
    "Banknifty": "retrieve_answer_bank",
    "Basics Of Stock Market": "retrieve_answer",
    "Out of Context": "retrieve_web_search"
})
graph_builder.add_edge("retrieve_answer", END)
graph_builder.add_edge("retrieve_answer_bank", END)
graph_builder.add_edge("retrieve_web_search", END)
# Compile graph
graph = graph_builder.compile()

image_bytes = graph.get_graph().draw_mermaid_png()

# Save the image bytes to a file
with open("flow.png", "wb") as f:
    f.write(image_bytes)

def process_query(user_input: str):
    state = {"query": user_input}
    for event in graph.stream(state):
        for value in event.values():
          if value in ['response'] :
            print("Assistant:", value["response"])

# Run chatbot in a loop
if __name__ == "__main__":
    while True:
        try:
            user_input = input("User: ")
            if user_input.lower() in ["quit", "exit", "q"]:
                print("Goodbye!")
                break
            process_query(user_input)
        except Exception as e:
            print(f"An error occurred: {e}")
            break
```
